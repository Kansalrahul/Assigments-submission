Ans:def compute_lcm(x,y):
    if x>y:
        greater=x
    else:
        greater=y
        
    while(True):
        if greater%x==0 and greater%y==0:

2. Write a Python Program to Find HCF?
Ans:def compute_hcf(a,b):
    if a>b:
        smaller=a
    else:
        smaller=b
        
        
    for  i in range(1,smaller+1):
        if a%i==0 and b%i==0:
            hcf=i
    return hcf

num1=int(input())
