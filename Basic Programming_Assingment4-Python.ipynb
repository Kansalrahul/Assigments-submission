1.Write a Python Program to Find the Factorial of a Number?
Ans:num=int(input())

factorial=1

if num<0:
    print('factorial number doesnt exist for negative number')
elif num==0:
    print('the factorial number is one')
    
else:
    for i in range(1,num+1):
        factorial=factorial*i
        print('the factorial of',num ,'is',factorial )

Ans:# Program to display the Fibonacci sequence up to n-th term

nterms = int(input("How many terms? "))


n1, n2 = 0, 1
count = 0

   print("Please enter a positive integer")

elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)

else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth=n1+n2
       # update values
       
       n1 = n2

4.Write a Python Program to Check Armstrong Number?
Ans:num=int(input())
sum=0
temp=num
while temp>0:
    
    digit=temp%10
    sum+=digit**3
    temp//=10
if num==sum:
    print(num,'this is armstrong number')
    
else:

    print(num,'this is not armstrong number')
